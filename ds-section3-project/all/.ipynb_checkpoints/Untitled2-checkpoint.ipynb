```python
import pandas as pd

df = pd.read_csv('merge.csv')
df.month = df.month.astype('str')
```

```python
(df.month == '8').sum()
```

```python
df_dum = pd.get_dummies(df, prefix=['month'])
```

```python
df_dum
```

```python
import numpy as np

skewed_data = ['pH','NTU','chl-a','cyan']

df_dum[skewed_data] = df_dum[skewed_data].applymap(lambda x: np.log(x) if x > 0 else 0)
```

```python
from sklearn.model_selection import train_test_split

target = 'cyan'
features = df_dum.columns.drop(target)

train, test = train_test_split(df_dum,test_size=0.2,random_state=2)
```

```python
X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

model = LinearRegression()

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)

print('테스트 데이터 MSE :',round(mean_squared_error(y_test,y_pred)))
print('테스트 데이터 RMSE :',round(mean_squared_error(y_test,y_pred)**0.5))
print('테스트 데이터 MAE :',round(mean_absolute_error(y_test,y_pred)))
print('테스트 데이터 r2_score :',r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_jobs=-1, random_state=2)

rf.fit(X_train,y_train)
y_pred_rf = rf.predict(X_test)

print('훈련 R2:',r2_score(y_train,rf.predict(X_train)))
print('테스트 데이터 MSE :',mean_squared_error(y_test,y_pred_rf))
print('테스트 데이터 RMSE :',mean_squared_error(y_test,y_pred_rf)**0.5)
print('테스트 데이터 MAE :',mean_absolute_error(y_test,y_pred_rf))
print('테스트 데이터 r2_score :',r2_score(y_test,y_pred_rf))

print('훈련 MSE:',mean_squared_error(y_train, rf.predict(X_train)))
print('테스트 MSE:',mean_squared_error(y_test, rf.predict(X_test)))
print('훈련 R2:',r2_score(y_train,rf.predict(X_train)))
print('테스트 R2:',r2_score(y_test,rf.predict(X_test)))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV

params = {'n_estimators' : randint(low=100, high=500),
          'max_depth' : randint(low=3, high=7),
          'max_features' : randint(low=1, high=10),
          'min_samples_leaf' : randint(low=1, high=15)
         }

rf = RandomForestRegressor(n_jobs=-1, random_state=2)

clf = RandomizedSearchCV(rf, params, random_state=2, cv=10)

clf.fit(X_train,y_train)
print('훈련 MSE:',mean_squared_error(y_train, clf.predict(X_train)))
print('테스트 MSE:',mean_squared_error(y_test, clf.predict(X_test)))
print('훈련 R2:',r2_score(y_train,clf.predict(X_train)))
print('테스트 R2:',r2_score(y_test,clf.predict(X_test)))
```
