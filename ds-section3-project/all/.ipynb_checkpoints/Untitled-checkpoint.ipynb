```python
import os
import pandas as pd
from glob import glob

# os.chdir(os.path.abspath(os.path.dirname(__file__)))

dfs = []

for file in glob('*.csv'):
    
    if file == 'merge.csv':
        continue

    df_ = pd.read_csv(file,encoding='cp949',parse_dates=[0])
    dfs.append(df_)

df = pd.concat(dfs)

df.columns = ['date','tavg','tmin','tmax','wt','pH','DO','NTU','chl-a','cyan']

df.drop(['date'],axis=1,inplace=True)

df.dropna(inplace=True)
```

```python
dfs2 = []

for file in glob('*.csv'):

    df_ = pd.read_csv(file,encoding='cp949',parse_dates=[0])
    dfs.append(df_)

df2 = pd.concat(dfs)

df2.columns = ['date','tavg','tmin','tmax','wt','pH','DO','tur','chl-a','cyan']

df2['month'] = df2.date.dt.month.astype('str')

df2.drop(['date'],axis=1,inplace=True)

df2.dropna(inplace=True)
```

```python
df_dum = pd.get_dummies(df2, prefix=['month'])
```

```python
df_dum
```

```python
skewed_data = ['pH','tur','chl-a','cyan']

df_dum[skewed_data] = df_dum[skewed_data].applymap(lambda x: np.log(x) if x > 0 else 0)
```

```python
df_dum
```

```python
target = 'cyan'
features = df_dum.columns.drop(target)

train, test = train_test_split(df_dum,test_size=0.2,random_state=2)

X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LinearRegression()

model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)

print('MAE =',round(mae))

y_pred = model.predict(X_train_scaled)

print('í›ˆë ¨ ë°ì´í„° MSE :',round(mean_squared_error(y_train,y_pred)))
print('í›ˆë ¨ ë°ì´í„° RMSE :',round(mean_squared_error(y_train,y_pred)**0.5))
print('í›ˆë ¨ ë°ì´í„° MAE :',round(mean_absolute_error(y_train,y_pred)))
print('í›ˆë ¨ ë°ì´í„° r2_score :',r2_score(y_train,y_pred))

y_pred = model.predict(X_test_scaled)

print('í…ŒìŠ¤íŠ¸ ë°ì´í„° MSE :',round(mean_squared_error(y_test,y_pred)))
print('í…ŒìŠ¤íŠ¸ ë°ì´í„° RMSE :',round(mean_squared_error(y_test,y_pred)**0.5))
print('í…ŒìŠ¤íŠ¸ ë°ì´í„° MAE :',round(mean_absolute_error(y_test,y_pred)))
print('í…ŒìŠ¤íŠ¸ ë°ì´í„° r2_score :',r2_score(y_test,y_pred))
```
